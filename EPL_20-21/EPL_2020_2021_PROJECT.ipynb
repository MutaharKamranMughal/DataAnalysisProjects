```python
# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# Load DataSet

epl_df = pd.read_csv('EPL_20_21.csv')
epl_df.head()
```

```python
# To Check Null and Non-Null Values

epl_df.info()
```

```python
# To Get Summary/Statistics

epl_df.describe()
```

```python
# To check total number of null values in the columns

epl_df.isna().sum()
```

```python
# Creating two new columns:

#First Column
epl_df['Mins_Per_Match'] = (epl_df['Mins'] / epl_df['Matches']).astype(int)

# Second Column
epl_df['Goals_Per_Match'] = (epl_df['Goals'] / epl_df['Matches']).astype(float)

epl_df
```

```python
# Total Goals in the whole season:

total_goals_season = epl_df['Goals'].sum()
print(f'total_goals_season = {total_goals_season}')
```

```python
# Total Goals by Penalties:

total_goals_penalties = epl_df['Penalty_Goals'].sum()
print(f'total_goals_penalties = {total_goals_penalties}')
```

```python
# Penalty Attempts: 
total_penalty_attempts = epl_df['Penalty_Attempted'].sum()
print(f'total_penalty_attempts = {total_penalty_attempts}')
```

```python
# Pie Chart for Penalties Scored Vs Missed:

# Penalties not Scored:
plt.figure(figsize=(13,6))
penalty_not_scored = total_penalty_attempts - total_goals_penalties
data = [penalty_not_scored, total_goals_penalties]
labels = ['Penalties Missed', 'Penalties Scored']
color = sns.color_palette('Set2')
plt.pie(data, labels=labels, colors=color, autopct='%.0f%%') # autopct gives us the format in which we can display our data.
plt.show()
```

```python
# Unique Player Positions:

epl_df['Position'].unique()
```

```python
# Total FW Players:

epl_df[epl_df['Position'] == 'FW'] # subsetting data frame to derive valuable info.
```

```python
# Players from Different Nations:

np.size((epl_df['Nationality'].unique()))
#epl_df[epl_df['Nationality'] != 'ENG']
```

```python
# Most Players from 'Which Nation?':

nationality = epl_df.groupby('Nationality').size().sort_values(ascending=False)
nationality.head(10).plot(kind='bar', figsize=(12,6), color=sns.color_palette('magma'))
```

```python
# Clubs with maximum number of players.

epl_df['Club'].value_counts().nlargest(10).plot(kind='bar', color=sns.color_palette('viridis'))  # '.vsalue_counts()'counts the values in the column.
```

```python
# Clubs with least number of players in the squad.

epl_df['Club'].value_counts().nsmallest(10).plot(kind='bar', color=sns.color_palette('viridis'))
```

```python
# Players based on age group.

under_20 = epl_df[epl_df['Age'] <= 20]
age20_25 = epl_df[(epl_df['Age'] > 20) & (epl_df['Age'] <= 25)]
age25_30 = epl_df[(epl_df['Age'] > 25) & (epl_df['Age'] <= 30)]
over_30 = epl_df[epl_df['Age'] > 30]
```

```python
# Creating pie chart based on age distribution through out the league.

x = np.array([under_20['Name'].count(), age20_25['Name'].count(), age25_30['Name'].count(), over_30['Name'].count()])
my_labels = ['<=20', '>20 & <=25', '>25 & <=30', '>30']
plt.title('Total Players with Different Age Groups', fontsize=15)
plt.pie(x, labels= my_labels, autopct='%.1f%%')
plt.show()
```

```python
# Totals number of under 20 players in each club

players_under_20 = epl_df[epl_df['Age'] < 20]
players_under_20['Club'].value_counts().plot(kind='bar', color=sns.color_palette('cubehelix'))
```

```python
# Number of players under 20 in MANU

players_under_20[players_under_20['Club'] == 'Manchester United']
```

```python
# Number of players under 20 in CHELSEA.

players_under_20[players_under_20['Club'] == 'Chelsea']
```

```python
# Average age of players in each club.

plt.figure(figsize=(12, 6))
sns.boxplot(x='Club', y='Age', data=epl_df)
plt.xticks(rotation=90)
```

```python
num_player = epl_df.groupby('Club').size()
data = (epl_df.groupby('Club')['Age'].sum()) / num_player
data.sort_values(ascending=False)
# print(num_player)
# print(data)
```

```python
# Total Assists from Each Clubs:

assists_by_clubs = pd.DataFrame(epl_df.groupby('Club', as_index=False)['Assists'].sum())
sns.set_theme(style='whitegrid', color_codes=True)
ax = sns.barplot(x='Club', y='Assists', data=assists_by_clubs.sort_values(by='Assists'), palette='Set2')
ax.set_xlabel('Club', fontsize=30)
ax.set_ylabel('Assists', fontsize=20)
plt.xticks(rotation=75)
plt.rcParams['figure.figsize'] = (20,8)
plt.title('Plot of Clubs vs Total Assists', fontsize=20)
```

```python
# Top 10 Assists

top_10_assists = epl_df[['Name', 'Club', 'Assists', 'Matches']].nlargest(n=10, columns='Assists')
top_10_assists
```

```python
# Total Goals by Each CLub:

goals_by_clubs = pd.DataFrame(epl_df.groupby('Club', as_index=False)['Goals'].sum())
sns.set_theme(style='whitegrid', color_codes=True)
ax = sns.barplot(x='Club', y='Goals', data=goals_by_clubs.sort_values(by='Goals'), palette='rocket')
ax.set_xlabel('Club', fontsize=30)
ax.set_ylabel('Goals', fontsize=20)
plt.xticks(rotation=75)
plt.rcParams['figure.figsize'] = (20,8)
plt.title('Plot of CLub vs Total Goals', fontsize=20)
```

```python
# Most Goals by Players:

top_10_goals = epl_df[['Name', 'Club', 'Goals', 'Matches']].nlargest(n=10, columns='Goals')
top_10_goals
```

```python
# Goals per Match:

top_10_goals_per_match = epl_df[['Name', 'Goals_Per_Match', 'Matches', 'Goals']].nlargest(n=10, columns='Goals_Per_Match')
top_10_goals_per_match
```

```python
# Pie Chart - Goals with assist and without assit:

plt.figure(figsize=(14,7))
assists = epl_df['Assists'].sum()
data = [total_goals_season - assists, assists]
labels = ['Goals without Assists', 'Goals with Assists']
color = sns.color_palette('Set1')
plt.pie(data, labels=labels, colors=color, autopct='%.0f%%')
plt.show
```

```python
# Top 10 players with Most Yellow Cards:

epl_yellow = epl_df.sort_values(by='Yellow_Cards', ascending=False)[:10]
plt.figure(figsize=(20,6))
plt.title('Players with the Most Yellow Cards')
c = sns.barplot(x=epl_yellow['Name'], y=epl_yellow['Yellow_Cards'], label= 'Players', color='yellow')
plt.ylabel('Number of Yellow Cards')
c.set_xticklabels(c.get_xticklabels(), rotation=45)
c
```
